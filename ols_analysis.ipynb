```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import xlrd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from matplotlib.pyplot import savefig
import os
data=pd.read_excel('C:\\Users\\Liwen Guo\\OneDrive - UNSW\\Yuan\\结果_0820\\data.xlsx',header=0,index_col=0)
data.head()
```

```python
model8 = sm.OLS(data['satis_overall'],sm.add_constant(data[['age', 'income']]))
result = model8.fit()

y=data['satis_overall']
x=data[['age', 'income']]
y_fitted = result.fittedvalues
fig, ols_satis_overall = plt.subplots(figsize=(8,6))
ols_satis_overall.plot(x,y, 'o', label = 'satis_overall')
ols_satis_overall.plot(x,y_fitted,'r--.',label='satis_overall_hat')
ols_satis_overall.legend(loc='best')
savefig('C:\\Users\\Liwen Guo\\OneDrive - UNSW\\Yuan\\结果_0820\\figure5_1')


result.summary()
print(result.summary())
out = result.summary()
with open('table5_1.csv','w') as f:
    f.write(out.as_csv())
```
